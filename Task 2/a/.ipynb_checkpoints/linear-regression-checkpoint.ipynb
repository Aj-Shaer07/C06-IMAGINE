# Data Loading

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")


print("Training Data:")
display(train_data.head())

print("Test Data:")
display(test_data.head())


print("\nTraining Data Info:")
train_data.info()

print("\nTest Data Info:")
test_data.info()
```

# Data Preprocessing

## Standardize the data
    Standardization is the process of transforming features so that they have a mean of 0 and a standard deviation of 1. This ensures that all             features contribute equally to the model and prevents features with larger scales from dominating the learning process.
### Why Use Standardization in Machine Learning?
 Standardization is important because many machine learning algorithms, including linear regression with gradient descent, are sensitive to the scale of input features. When features are on different scales, the optimization process can become slow and unstable. Standardization helps in:
- Faster convergence of gradient descent.
- Improved numerical stability.
- Ensuring that each feature contributes equally to the model.
- Better model performance and interpretability.

### How to Standardize Data

    Data can be standardized using the z-score formula:

z = (x - μ) / σ

where:
- x is the original feature value,
- μ is the mean of the feature,
- σ is the standard deviation of the feature.

In practice, we compute the mean and standard deviation from the training data and use these values to transform both the training and test datasets.

```python
# Extract feature and target
X_train = train_data.iloc[:, 0].values
y_train = train_data.iloc[:, 1].values
X_test = test_data.iloc[:, 0].values
y_test = test_data.iloc[:, 1].values

# Compute mean and standard deviation from training data
mean_X = np.mean(X_train)
std_X = np.std(X_train)

# Standardize the data
X_train = (X_train - mean_X) / std_X
X_test = (X_test - mean_X) / std_X

print("Data standardized successfully.")
```

## Reshaping data for the correct shape for the model


Most machine learning models, including our custom linear regression implementation, expect the input data to be in a two-dimensional format of shape (number of samples, number of features). If the data is one-dimensional, matrix operations such as dot products and broadcasting may fail or produce incorrect results. Reshaping ensures compatibility with vectorized computations and allows efficient implementation of gradient descent.

```python
# Ensure correct shapes
X_train = X_train.reshape(-1, 1) if X_train.ndim == 1 else X_train
X_test = X_test.reshape(-1, 1) if X_test.ndim == 1 else X_test
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

# Model Implementation

```python
class LinearRegression:
    
    def __init__(self, learning_rate=0.01):
        self.learning_rate = learning_rate
        self.w = None
        self.b = None
        self.cost_history = []

    def initialize_parameters(self, n_features):
        self.w = np.zeros((n_features, 1))
        self.b = 0.0

    def forward(self, X):
        return np.dot(X, self.w) + self.b

    def compute_cost(self, y_pred, y):
        m = y.shape[0]
        cost = (1 / (2 * m)) * np.sum((y_pred - y) ** 2)
        return cost

    def backward(self, X, y, y_pred):
        m = X.shape[0]
        dw = (1 / m) * np.dot(X.T, (y_pred - y))
        db = (1 / m) * np.sum(y_pred - y)
        return dw, db

    def fit(self, X, y, iterations=1000, plot_cost=True):
        m, n = X.shape
        self.initialize_parameters(n)

        for i in range(iterations):
            y_pred = self.forward(X)
            cost = self.compute_cost(y_pred, y)
            dw, db = self.backward(X, y, y_pred)

            self.w -= self.learning_rate * dw
            self.b -= self.learning_rate * db

            self.cost_history.append(cost)

            if i % 100 == 0:
                print(f"Iteration {i}: Cost = {cost}")

        if plot_cost:
            import matplotlib.pyplot as plt
            plt.plot(self.cost_history)
            plt.xlabel("Iterations")
            plt.ylabel("Cost")
            plt.title("Cost Reduction Over Time")
            plt.show()

    def predict(self, X):
        return self.forward(X)

    def save_model(self, filename='model.pkl'):
        with open(filename, 'wb') as f:
            pickle.dump({'w': self.w, 'b': self.b}, f)

    @classmethod
    def load_model(cls, filename):
        with open(filename, 'rb') as f:
            params = pickle.load(f)
        model = cls()
        model.w = params['w']
        model.b = params['b']
        return model
   
```

```python
lr = LinearRegression(learning_rate=0.01)
lr.fit(X_train, y_train, iterations=1000)

# Use the trained model
model = lr
```

```python
lr.save_model('model.pkl')
```

# Evaluation



### 1. Mean Squared Error (MSE)

**Formula:**
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_{\text{true}_i} - y_{\text{pred}_i})^2
$$

**Description:**
 - Mean Squared Error (MSE) measures the average of the squared differences between the actual and predicted values. It penalizes larger errors more heavily due to the squaring operation.

**Interpretation:**
-  A lower MSE indicates better model performance. Since the errors are squared, the metric is sensitive to outliers and is expressed in squared units of the target variable.

### 2. Root Mean Squared Error (RMSE)

**Formula:**
$$
\text{RMSE} = \sqrt{\text{MSE}}
$$

**Description:**
- Root Mean Squared Error (RMSE) is the square root of the Mean Squared Error. It provides an error metric in the same units as the target variable, making it easier to interpret.
**Interpretation:**
- A lower RMSE indicates better predictive accuracy. Because it is expressed in the same units as the target variable, it represents the typical magnitude of prediction errors.

### 3. R-squared ($R^2$)

**Formula:**
$$
R^2 = 1 - \frac{\text{SSR}}{\text{SST}}
$$

**Description:**
 R-squared (Coefficient of Determination) measures the proportion of the variance in the dependent variable that is explained by the independent variables in the model.

**Interpretation:**
 R² values range from 0 to 1. A value close to 1 indicates that the model explains most of the variability in the target variable, while a value close to 0 suggests poor explanatory power. Negative values may occur if the model performs worse than simply predicting the mean.

```python
class RegressionMetrics:
    @staticmethod
    def mean_squared_error(y_true, y_pred):
        y_true = y_true.reshape(-1, 1)
        y_pred = y_pred.reshape(-1, 1)
        return np.mean((y_true - y_pred) ** 2)

    @staticmethod
    def root_mean_squared_error(y_true, y_pred):
        mse = RegressionMetrics.mean_squared_error(y_true, y_pred)
        return np.sqrt(mse)

    @staticmethod
    def r_squared(y_true, y_pred):
        y_true = y_true.reshape(-1, 1)
        y_pred = y_pred.reshape(-1, 1)
        ss_res = np.sum((y_true - y_pred) ** 2)
        ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
        return 1 - (ss_res / ss_tot)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)

# Plot predictions vs actual values
fig = px.scatter(
    x=X_test.flatten(),
    y=y_test.flatten(),
    title="Actual vs Predicted Values",
    labels={"x": "Feature", "y": "Target"}
)
fig.add_scatter(
    x=X_test.flatten(),
    y=y_pred.flatten(),
    mode='lines',
    name='Regression Line'
)
fig.show()
```

```python
y_pred = model.predict(X_test)
mse_value = RegressionMetrics.mean_squared_error(y_test, y_pred)
rmse_value = RegressionMetrics.root_mean_squared_error(y_test, y_pred)
r_squared_value = RegressionMetrics.r_squared(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse_value}")
print(f"Root Mean Squared Error (RMSE): {rmse_value}")
print(f"R-squared (Coefficient of Determination): {r_squared_value}")
```
