```python
data=[1,2,4,5,7]
def normalize(data):
    min_val = min(data)
    max_val = max(data)
    
    # Avoid division by zero
    if max_val == min_val:
        return [0 for _ in data]
    
    normalized = [(x - min_val) / (max_val - min_val) for x in data]
    return normalized
```
